```python
import torch
from sklearn.model_selection import train_test_split
import torch.nn as nn
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
data = pd.read_csv("./datasets/housing_v2.csv")
```

```python
data.info()
```

```python
# perform in place operation to drop nan rows
data.dropna(inplace=True)
```

```python
data.info()
```

```python
# split into test/train
from sklearn.model_selection import train_test_split
```

```python
# ft engineering
def process_data(data):
    data['total_rooms'] = np.log(data["total_rooms"] + 1)
    data['total_bedrooms'] = np.log(data["total_bedrooms"] + 1)
    data['population'] = np.log(data["population"] + 1)
    data['households'] = np.log(data["households"] + 1)
    onehot = pd.get_dummies(data.ocean_proximity)
    data = data.join(onehot)
    data = data.drop(['ocean_proximity'], axis=1)
    data['bedroom_ratio'] = data['total_bedrooms'] / data['total_rooms']
    data['household_rooms'] = data['total_rooms'] / data['households']
    return data
```

```python
data = process_data(data)
features = data.drop("median_house_value", axis=1)
targets = data["median_house_value"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, targets, test_size=0.3)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.score(x_test, y_test)
```

```python
# try scaling the data
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

x_train_s = scaler.fit_transform(x_train)
x_test_s = scaler.transform(x_test)
reg.fit(x_train_s, y_train)
```

```python
reg.score(x_test_s, y_test)
```

```python
# random forest
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(x_train, y_train)
forest.score(x_test, y_test)
```

```python
# try with scaling
forest.fit(x_train_s, y_train)
forest.score(x_test_s, y_test)
```

```python
# try with cross validation
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators':[2, 4, 4],
    'max_features': [8, 12, 20],
    'min_samples_split':[2, 4, 6, 8]
}

grid_search = GridSearchCV(forest, param_grid, 
                           cv=5, 
                           scoring='neg_mean_squared_error', 
                           return_train_score=True)
grid_search.fit(x_train_s, y_train)
```

```python
# get the best, optimal hyper params
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(x_test_s, y_test)
```
